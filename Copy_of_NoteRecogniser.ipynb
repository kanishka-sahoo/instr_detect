```python
import librosa, glob, os

musicdata = pd.DataFrame()
path = 'audio'
i=0
for filename in glob.glob(os.path.join(path, '*.wav')):
    y, sr = librosa.load(filename, sr=16000)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    name = filename[:]
    startindex = name.index('\\')+1
    endindex = name.index('.')
    name = name[startindex:endindex]
    singlecolumn = pd.DataFrame()
    singlecolumn[name] = [mfcc, sr]
    #musicdata.insert(0, name, [y, sr], True)
    musicdata = pd.concat((musicdata, singlecolumn), axis = 1)
    i+=1
    if i == 50:
      break

musicdata.head()
```

```python
X = df['y']
X.head()
```

```python
y = pd.get_dummies(df['instrument_family_str'])
y.head()
```

```python
import tensorflow
import keras

from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.20)

model = Sequential()
model.add(layers.Conv2D(16, (3, 3), activation='relu', input_shape=(16, 13, 162)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPool2D((2, 2)))
model.add(layers.Dropout(0.5))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.summary()
```

```python
history = model.fit(train_x, train_y, epochs=30, validation_data=(test_x, test_y))
```
